- Haz un histograma de la distribución de las magnitudes de los terremoto

```python
# Descargue los datos ejecutando la siguiente instrucción
!curl -L -o earthquakes.csv "https://risk-engineering.org/static/data/earthquakes-2017.csv"
```

```python
import pandas as pd

data = pd.read_csv('earthquakes.csv', index_col=['time'], parse_dates=['time'])
data.tail()
```

```python
import matplotlib.pyplot as plt

# Argupamos los datos de acuerdo al timpo

new_data = data.resample('1m').mean()

new_data.head()
```

```python
plt.plot(new_data.index, new_data['mag'], 'o')
plt.xlabel("Tiempo")
plt.ylabel("Magnitud")
plt.show()
```

```python
plt.hist(data.mag.dropna())
plt.xlabel("Magnitud")
plt.ylabel("Frecuencia")
plt.show()
```

### Problema 2 (4 Pts)

Suponga que un vendedor de frutas vendió 20 mangos y 10 naranjas en un día por un total de $350. Al día siguiente vendió 17 mangos y 22 naranjas por $500. Si los precios de las frutas se mantuvieron sin cambios en ambos días, ¿cuál fue el precio de un mango y una naranja? Utilice linalg para resolver este problema. 

```python
import numpy as np

coef = np.array([[20, 10], [17,  22]])
result = np.array([350, 500])

respuesta = np.linalg.solve(coef, result)
respuesta
```

Respuesta: Los mangos valen 10 USD cada uno y las naranjas 15 USD
